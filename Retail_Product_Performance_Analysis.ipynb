# Retail Product Performance Analysis

## Business Problem
A retail company wants to analyze product sales performance to identify:

- Top selling products
- Best performing categories
- Revenue contribution by product
- Overall sales distribution

The goal is to generate insights that help improve product strategy.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "product": [
        "Laptop","Laptop","Laptop",
        "Phone","Phone","Phone",
        "Headphones","Headphones",
        "Keyboard","Keyboard"
    ],

    "category": [
        "Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics",
        "Accessories","Accessories",
        "Accessories","Accessories"
    ],

    "price": [1000,1000,1000,700,700,700,150,150,80,80],

    "units_sold": [5,3,4,8,6,7,10,12,9,11]
}

df = pd.DataFrame(data)
df
```

```python
#creating a Revenue coloumn
df['revenue'] = df['price'] * df['units_sold']
df

df.groupby('product')['revenue'].sum()
```

```python
df.groupby('category')['revenue'].sum()
```

```python
df.groupby('product')['revenue'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Revenue by product')
plt.xlabel('product')
plt.ylabel('Total Revenue')
plt.show()
```
